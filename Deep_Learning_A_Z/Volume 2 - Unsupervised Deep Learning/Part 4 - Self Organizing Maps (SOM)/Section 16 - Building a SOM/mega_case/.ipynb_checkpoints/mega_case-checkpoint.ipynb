## 1. Identify the Frauds with the Self-Organizing Map

```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from minisom import MiniSom
from pylab import bone, pcolor, colorbar, plot, show
%matplotlib inline
```

```python
dataset = pd.read_csv("Credit_Card_Applications.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
sc = MinMaxScaler(feature_range = (0,1))
X_norm = sc.fit_transform(X)
som = MiniSom(x = 10, y = 10, input_len=15, sigma=1.0, learning_rate=0.5)
som.random_weights_init(X)
som.train_random(data = X, num_iteration=100)
```

```python
bone() # Initiliaze the plot
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']
colors = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0]+0.5,
         w[1]+0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = "None",
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mappings = som.win_map(X)
frauds = np.concatenate((mappings[(4,5)], mappings[(5,4)]), axis = 0)
frauds = sc.inverse_transform(frauds)
frauds
```

## 2. Going from Unsupervised to Supervised Deep Learning

Creating the matrix of features

```python
customers = dataset.iloc[:, 1:].values
```

Creating the dependent variable

```python
is_fraud = np.zeros(len(dataset))
```

```python
for i in range(len(dataset)):
    if dataset.iloc[i, 0] in frauds:
        is_fraud[i] = 1
```

```python
print(is_fraud)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
customers = sc.fit_transform(customers)

# Part 2 - Now let's make the ANN!

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 2, kernel_initializer = 'uniform', activation = 'relu', input_dim = 15))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(customers, is_fraud, batch_size = 1, epochs = 1)

# Part 3 - Making predictions and evaluating the model

```

```python
# Predicting the Test set results
y_pred = classifier.predict(customers)
```

```python
print(y_pred)
```

```python
y_pred = np.concatenate((dataset.iloc[:,0:1].values,y_pred), axis = 1)
```

```python
y_pred = y_pred[y_pred[:,1].argsort()]
```
