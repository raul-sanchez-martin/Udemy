```python
import nltk
import random
```

```python
from nltk.corpus import names
```

```python
names.fileids()
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use("ggplot")
```

```python
name_cfd = nltk.ConditionalFreqDist((fileid, name[-2:])for fileid in names.fileids() 
                                    for name in names.words(fileid))
```

```python
plt.figure(figsize=(50,10))
name_cfd.plot()
```

```python
def name_features(name):
    return {"pair" : name[-2:]}
```

```python
name_features("Katy")
```

```python
name_list = [(name, "male") for name in names.words("male.txt")] + [(name, "female") for name in names.words("female.txt")]
```

```python
name_list[:10]
```

```python
name_list[:-10]
```

```python
random.shuffle(name_list)
```

```python
name_list[:10]
```

```python
features = [(name_features(name),gender) for (name,gender) in name_list]
```

```python
features[:10]
```

```python
len(features)/2
```

```python
training_set = features[:3972]
testing_set = features[3972:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
male_names = names.words("male.txt")
```

```python
"Carmello" in male_names
```

```python
classifier.classify(name_features("Carmello"))
```

```python
nltk.classify.accuracy(classifier, testing_set)
```
