Find all the parts of speech for the words over, spoke and answer

```python
import nltk
```

```python
alice = nltk.corpus.gutenberg.words("carroll-alice.txt")
```

```python
alice_norms = [word.lower() for word in alice if word.isalpha()]
```

```python
alice_tags = nltk.pos_tag(alice_norms, tagset="universal")
```

```python
alice_cfd = nltk.ConditionalFreqDist(alice_tags)
```

```python
alice_cfd
```

```python
alice_cfd["over"]
```

```python
alice_cfd["spoke"]
```

```python
alice_cfd["answer"]
```
